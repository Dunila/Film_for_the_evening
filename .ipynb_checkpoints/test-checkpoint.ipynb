# Parsing first 1000 IDies

```python
films = []
```

```python
for i in range(360, 500):
    try:
        driver.get(f"https://www.kinopoisk.ru/film/{i}/")
        WebDriverWait(driver, 5).until(
            EC.presence_of_element_located((By.XPATH, "//h1[@itemprop='name']//span"))
        )

        if "проверка безопасности" in driver.page_source.lower():
            print("🛑 Капча! Останавливаемся.")
            break

        film_info = parse_film(i, driver=driver)
        if film_info:
            films.append(film_info)
            print(f"✅ Parsed film {i}")
        else:
            print(f"⚠️ Skipped film {i}")

    except Exception as e:
        print(f"❌ Error with film {i}: {e}")
    
    time.sleep(np.random.uniform(0.5, 3))
```
